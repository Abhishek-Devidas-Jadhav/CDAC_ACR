### split the data into x and y

```python
x = df.drop(selected_feature, axis=1)
y = df[selected_feature]
```

### split the data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=123456)
```

### model training

```python
def train_model_lg():
    from sklearn.linear_model import LogisticRegressionCV

    model = LogisticRegressionCV(max_iter=10000)
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_knn():
    from sklearn.neighbors import KNeighborsClassifier

    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_svm():
    from sklearn.svm import SVC

    model = SVC(C=2.0, gamma='auto')
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_nb():
    from sklearn.naive_bayes import GaussianNB

    model = GaussianNB()
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_dt():
    from sklearn.tree import DecisionTreeClassifier

    model = DecisionTreeClassifier(max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_rf():
    from sklearn.ensemble import RandomForestClassifier

    model = RandomForestClassifier(n_estimators=100)
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_catboost():
    from catboost import CatBoostClassifier

    model = CatBoostClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_adaboot():
    from sklearn.ensemble import AdaBoostClassifier

    model = AdaBoostClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def train_model_xgb():
    from sklearn.preprocessing import LabelEncoder
    
    le = LabelEncoder()
    y_train_temp = le.fit_transform(y_train)
    
    from xgboost import XGBClassifier

    model = XGBClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
models = [
   #(train_model_lg(), 'Logistic Regression'),
    (train_model_knn(), 'KNN'),
    (train_model_svm(), 'SVM'),
    (train_model_nb(), 'Naive Bayes'),
    (train_model_dt(), 'Decision Tree'),
    (train_model_rf(), 'Random Forest'),
    (train_model_catboost(), 'CatBoost'),
    (train_model_adaboot(), 'AdaBoost')]
"""
    (train_model_xgb(), 'XG Boost')
]
"""
```

### evalute the model

```python
def evaluate_model(model, model_name):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred,average='weighted')
    recall = recall_score(y_true, y_pred,average='weighted')
    f1 = f1_score(y_true, y_pred,average='weighted')

    return model_name, accuracy, precision, recall, f1
```

```python
results = []
for (model, model_name) in models:
    results.append(evaluate_model(model, model_name))
result_df = pd.DataFrame(results, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
result_df
```
